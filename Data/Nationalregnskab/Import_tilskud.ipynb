```python
params = {
    'table': 'TILSKUD',
    'format': 'BULK',
    'variables': [
        {'code': 'TYPE', 'values': ['010','085','090', '095']},
        {'code': 'TID', 'values': ['*']}
    ]
}
r = requests.post('https://api.statbank.dk/v1' + '/data', json=params)
print(r.text[:200])

df_f2015 = pd.read_csv(
    StringIO(
        requests.post('https://api.statbank.dk/v1' + '/data', json=params).text
    ), sep=';'
)
print(df_f2015.columns.tolist())
print(df_f2015.head())
```

```python
params1 = {
    'table': 'TILSKUD1',
    'format': 'BULK',
    'variables': [
        {'code': 'TILSKUDSART', 'values': ['170','175']},
        {'code': 'TID', 'values': ['*']}
    ]
}
r = requests.post('https://api.statbank.dk/v1' + '/data', json=params1)
print(r.text[:200])

df_2015 = pd.read_csv(
    StringIO(
        requests.post('https://api.statbank.dk/v1' + '/data', json=params1).text
    ), sep=';'
)
print(df_2015.columns.tolist())
print(df_2015.head())
```

```python
# 1) Sørg for at INDHOLD er tal (DK-format: komma som decimal)
df_f2015["INDHOLD"] = (
    df_f2015["INDHOLD"].astype(str)
      .str.replace("\u00a0", "", regex=False)  # non-breaking space
      .str.replace(" ", "", regex=False)       # mellemrum som tusindtals-sep
      .str.replace(".", "", regex=False)       # punktum som tusindtals-sep (hvis findes)
      .str.replace(",", ".", regex=False)      # komma -> punktum (decimal)
)
df_f2015["INDHOLD"] = pd.to_numeric(df_f2015["INDHOLD"], errors="coerce")
df_2015["INDHOLD"] = (
    df_2015["INDHOLD"].astype(str)
      .str.replace("\u00a0", "", regex=False)  # non-breaking space
      .str.replace(" ", "", regex=False)       # mellemrum som tusindtals-sep
      .str.replace(".", "", regex=False)       # punktum som tusindtals-sep (hvis findes)
      .str.replace(",", ".", regex=False)      # komma -> punktum (decimal)
)
df_2015["INDHOLD"] = pd.to_numeric(df_2015["INDHOLD"], errors="coerce")

```

```python
#Trækker braklægning og græs fra
df_f2005_hek = df_f2015[df_f2015['TYPE'] == 'Hektarstøtte i alt'].copy().fillna(0)
df_f2005_hek.drop(columns=['TYPE'], inplace=True)
df_f2005_brak = df_f2015[df_f2015['TYPE'] == 'Braklægningsstøtte'].copy().fillna(0)
df_f2005_brak.drop(columns=['TYPE'], inplace=True)
df_f2005_ekst = df_f2015[df_f2015['TYPE'] == 'Ekstensiveringspræmie'].copy().fillna(0)
df_f2005_ekst.drop(columns=['TYPE'], inplace=True)
df_2005_2015 = df_f2015[df_f2015['TYPE'] == 'Enkeltbetalingsordning, arealtilskud'].copy().fillna(0)
df_2005_2015.drop(columns=['TYPE'], inplace=True)
df_2015_grundbetaling = df_2015[df_2015['TILSKUDSART'] == '2.1 Grundbetaling'].copy().fillna(0)
df_2015_grundbetaling.drop(columns=['TILSKUDSART'], inplace=True)
df_2015_gron = df_2015[df_2015['TILSKUDSART'] == '2.2 Grønne krav'].copy().fillna(0)
df_2015_gron.drop(columns=['TILSKUDSART'], inplace=True)

df_f2005_hek = df_f2005_hek.set_index(['TID'])
df_f2005_brak = df_f2005_brak.set_index(['TID'])
df_f2005_ekst = df_f2005_ekst.set_index(['TID'])
df_2005_2015 = df_2005_2015.set_index(['TID'])
df_2015_grundbetaling = df_2015_grundbetaling.set_index(['TID'])
df_2015_gron = df_2015_gron.set_index(['TID'])  

df_f2005 = df_f2005_hek['INDHOLD'] + df_f2005_brak['INDHOLD'] + df_f2005_ekst['INDHOLD']
df_f2005 = df_f2005.reset_index(name='INDHOLD')
df_f2005 = df_f2005[df_f2005['TID'] < 2005].copy()
df_2005_2015 = df_2005_2015['INDHOLD'].reset_index(name='INDHOLD')
df_2005_2015 = df_2005_2015[(df_2005_2015['TID'] >= 2005) & (df_2005_2015['TID'] < 2015)].copy()
df_e2015 = df_2015_grundbetaling['INDHOLD'] + df_2015_gron['INDHOLD']
df_e2015 = df_e2015.reset_index(name='INDHOLD')
df_e2015 = df_e2015[df_e2015['TID'] >= 2015].copy()
df_f2005
df_2005_2015
df_e2015

df_tilskud_samlet = pd.concat([df_f2005, df_2005_2015, df_e2015], ignore_index=True)

# Sorter efter TID for at få dem i kronologisk rækkefølge
df_tilskud_samlet = df_tilskud_samlet.sort_values('TID').reset_index(drop=True)
df_tilskud_samlet
df_tilskud_samlet.to_csv('Data69/landbrugsdata_tilskud.csv', index=False)
```
